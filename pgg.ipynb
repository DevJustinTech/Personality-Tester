```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# Identify categorical and numerical columns
categorical_cols = train_data.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_cols = train_data.select_dtypes(include=['int64', 'float64']).columns.tolist()
if 'Personality' in categorical_cols:
    categorical_cols.remove('Personality')  # Remove target if present
if 'Personality' in numerical_cols:
    numerical_cols.remove('Personality')  # Remove target if present

# Define transformers
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))
])

# Combine transformers into a preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ]
 )

# Fit the preprocessor on train data and transform both train and test
X_train = train_data.drop('Personality', axis=1)
y_train = train_data['Personality']
X_test = test_data.copy()
X_train_prepared = preprocessor.fit_transform(X_train)
X_test_prepared = preprocessor.transform(X_test)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_curve, auc

# Split the prepared data for validation
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train_prepared, y_train, test_size=0.2, random_state=42
)

# Scale features on the split data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_split)
X_val_scaled = scaler.transform(X_val_split)

# Handle class imbalance with SMOTE (use scaled training data)
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train_split)

# Retrain and evaluate model
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_balanced, y_train_balanced)
y_val_pred = rf.predict(X_val_scaled)
y_val_proba = rf.predict_proba(X_val_scaled)[:, 1] if len(rf.classes_) == 2 else None

print('Classification Report:')
print(classification_report(y_val_split, y_val_pred))

if y_val_proba is not None:
    fpr, tpr, thresholds = roc_curve(y_val_split.map({'Extrovert': 1, 'Introvert': 0}), y_val_proba)
    roc_auc = auc(fpr, tpr)
    print('ROC AUC Score:', roc_auc)
```

```python
# 1. Check class distribution in validation set
print('Validation set class distribution:')
print(y_val_split.value_counts())

# 2. Plot predicted probabilities
import matplotlib.pyplot as plt
plt.hist(y_val_proba, bins=50)
plt.xlabel('Predicted probability for Extrovert')
plt.ylabel('Frequency')
plt.title('Distribution of predicted probabilities')
plt.show()

# 3. Try Logistic Regression for comparison
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=42, max_iter=1000)
lr.fit(X_train_balanced, y_train_balanced)
y_val_proba_lr = lr.predict_proba(X_val_scaled)[:, 1]
print('Logistic Regression ROC AUC:', auc(*roc_curve(y_val_split.map({'Extrovert': 1, 'Introvert': 0}), y_val_proba_lr)[:2]))

# 4. Tune threshold for classification
import numpy as np
thresholds = np.linspace(0, 1, 100)
f1_scores = []
from sklearn.metrics import f1_score
for thresh in thresholds:
    y_pred_thresh = (y_val_proba_lr > thresh).astype(int)
    f1 = f1_score(y_val_split.map({'Extrovert': 1, 'Introvert': 0}), y_pred_thresh)
    f1_scores.append(f1)
best_thresh = thresholds[np.argmax(f1_scores)]
print('Best threshold for F1:', best_thresh)
print('Best F1 score:', max(f1_scores))
```

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# 1. Define pipeline
final_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('scaler', scaler),
    ('smote', SMOTE(random_state=42)),
    ('model', LGBMClassifier(random_state=42))
])

# 2. Define parameter grid for LightGBM
param_grid = {
    'model__n_estimators': [100, 200, 500],
    'model__max_depth': [-1, 10, 20],
    'model__learning_rate': [0.01, 0.05, 0.1],
    'model__num_leaves': [31, 50, 100],
    'model__min_child_samples': [10, 20, 30]
}

# 3. Setup StratifiedKFold for better class balance in CV
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 4. Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=final_pipeline,
    param_grid=param_grid,
    scoring='f1_macro',   # use 'accuracy', 'roc_auc', etc. if preferred
    cv=cv,
    n_jobs=-1,
    verbose=2
)

# 5. Fit on the raw training data
X_train = train_data.drop('Personality', axis=1)
y_train = train_data['Personality']

grid_search.fit(X_train, y_train)

# 6. Display best params and score
print(" Best Parameters:", grid_search.best_params_)
print(" Best Score:", grid_search.best_score_)
```

```python
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 1. Get best model from GridSearchCV
best_model = grid_search.best_estimator_

# 2. Create train-validation split
X = train_data.drop('Personality', axis=1)
y = train_data['Personality']

X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 3. Fit on training split and predict on validation split
best_model.fit(X_train_split, y_train_split)
y_val_pred = best_model.predict(X_val_split)

# 4. Classification Report
print("Classification Report:")
print(classification_report(y_val_split, y_val_pred))

# 5. Compute ROC AUC (works best for binary classification)
if len(y.unique()) == 2:
    # Map labels to 0 and 1 for binary ROC
    y_val_binary = y_val_split.map({'Extrovert': 0, 'Introvert': 1})
    
    # Get probability for positive class (Introvert = 1)
    positive_class_index = list(best_model.named_steps['model'].classes_).index('Introvert')
    y_val_proba = best_model.predict_proba(X_val_split)[:, positive_class_index]

    # ROC AUC Score
    roc_auc = roc_auc_score(y_val_binary, y_val_proba)
    print("ROC AUC Score:", roc_auc)

    # ROC Curve
    fpr, tpr, _ = roc_curve(y_val_binary, y_val_proba)
    plt.figure(figsize=(8,6))
    plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.3f})")
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve for Tuned LightGBM Model')
    plt.legend()
    plt.grid(True)
    plt.show()

# 6. Confusion Matrix
ConfusionMatrixDisplay.from_estimator(best_model, X_val_split, y_val_split)
plt.show()
```
